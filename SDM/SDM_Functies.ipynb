```python
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
import os

db_path = "databases/"

sdm_engine = create_engine(f'sqlite:///{db_path}BikeToDrive_SDM.db')
```

```python
e1 = create_engine(f'sqlite:///{db_path}BikeToDrive_1_Accessoireverkoop.db')
e2 = create_engine(f'sqlite:///{db_path}BikeToDrive_2_Fietsverkoop.db')
e3 = create_engine(f'sqlite:///{db_path}BikeToDrive_3_Onderhoud.db')
e4 = create_engine(f'sqlite:///{db_path}BikeToDrive_4_Accessoire_Inkoop.db')
e5 = create_engine(f'sqlite:///{db_path}BikeToDrive_5_Fiets_Inkoop.db')
# --- 1. Accessoireverkoop ---
Accessoireverkoop_Leverancier = pd.read_sql("SELECT * FROM Leverancier", e1)
Accessoireverkoop_Accessoire = pd.read_sql("SELECT * FROM Accessoire", e1)
Accessoireverkoop_Filiaal = pd.read_sql("SELECT * FROM Filiaal", e1)
Accessoireverkoop_Monteur = pd.read_sql("SELECT * FROM Monteur", e1)
Accessoireverkoop_Klant = pd.read_sql("SELECT * FROM Klant", e1)
Accessoire_Verkoop = pd.read_sql("SELECT * FROM Accessoire_Verkoop", e1)

# --- 2. Fietsverkoop ---
Fietsverkoop_Fabrikant = pd.read_sql("SELECT * FROM Fabrikant", e2)
Fietsverkoop_Fiets = pd.read_sql("SELECT * FROM Fiets", e2)
Fietsverkoop_Filiaal = pd.read_sql("SELECT * FROM Filiaal", e2)
Fietsverkoop_Monteur = pd.read_sql("SELECT * FROM Monteur", e2)
Fietsverkoop_Klant = pd.read_sql("SELECT * FROM Klant", e2)
Fiets_Verkoop = pd.read_sql("SELECT * FROM Fiets_Verkoop", e2)

# --- 3. Onderhoud ---
Onderhoud_Fabrikant = pd.read_sql("SELECT * FROM Fabrikant", e3)
Onderhoud_Fiets = pd.read_sql("SELECT * FROM Fiets", e3)
Onderhoud_Filiaal = pd.read_sql("SELECT * FROM Filiaal", e3)
Onderhoud_Monteur = pd.read_sql("SELECT * FROM Monteur", e3)
Onderhoud = pd.read_sql("SELECT * FROM Onderhoud", e3)

# --- 4. Accessoire Inkoop ---
Accessoire_Inkoop_Leverancier = pd.read_sql("SELECT * FROM Leverancier", e4)
Accessoire_Inkoop_Accessoire = pd.read_sql("SELECT * FROM Accessoire", e4)
Accessoire_Inkoop = pd.read_sql("SELECT * FROM Accessoire_Inkoop", e4)

# --- 5. Fiets Inkoop ---
Fiets_Inkoop_Fabrikant = pd.read_sql("SELECT * FROM Fabrikant", e5)
Fiets_Inkoop_Fiets = pd.read_sql("SELECT * FROM Fiets", e5)
Fiets_Inkoop = pd.read_sql("SELECT * FROM Fiets_Inkoop", e5)
```

```python
export_lijst = [
    (Accessoireverkoop_Leverancier, 'Accessoireverkoop_Leverancier'),
    (Accessoireverkoop_Filiaal, 'Accessoireverkoop_Filiaal'),
    (Accessoireverkoop_Klant, 'Accessoireverkoop_Klant'),
    (Fietsverkoop_Fabrikant, 'Fietsverkoop_Fabrikant'),
    (Fietsverkoop_Filiaal, 'Fietsverkoop_Filiaal'),
    (Fietsverkoop_Klant, 'Fietsverkoop_Klant'),
    (Onderhoud_Fabrikant, 'Onderhoud_Fabrikant'),
    (Onderhoud_Filiaal, 'Onderhoud_Filiaal'),
    (Accessoire_Inkoop_Leverancier, 'Accessoire_Inkoop_Leverancier'),
    (Fiets_Inkoop_Fabrikant, 'Fiets_Inkoop_Fabrikant'),
    (Accessoireverkoop_Accessoire, 'Accessoireverkoop_Accessoire'),
    (Accessoireverkoop_Monteur, 'Accessoireverkoop_Monteur'),
    (Fietsverkoop_Fiets, 'Fietsverkoop_Fiets'),
    (Fietsverkoop_Monteur, 'Fietsverkoop_Monteur'),
    (Onderhoud_Fiets, 'Onderhoud_Fiets'),
    (Onderhoud_Monteur, 'Onderhoud_Monteur'),
    (Accessoire_Inkoop_Accessoire, 'Accessoire_Inkoop_Accessoire'),
    (Fiets_Inkoop_Fiets, 'Fiets_Inkoop_Fiets'),
    (Accessoire_Verkoop, 'Accessoire_Verkoop'),
    (Fiets_Verkoop, 'Fiets_Verkoop'),
    (Onderhoud, 'Onderhoud'),
    (Accessoire_Inkoop, 'Accessoire_Inkoop'),
    (Fiets_Inkoop, 'Fiets_Inkoop')
]
```

```python
from sqlalchemy import text

with sdm_engine.begin() as conn:
    print("Start de Full-Refresh op het SDM")

    for df, tabelnaam in export_lijst:
        try:
            conn.execute(text(f"DELETE FROM {tabelnaam}"))

            df.to_sql(tabelnaam, con=conn, if_exists='append', index=False)

            print(f"{tabelnaam}: {len(df)} rijen succesvol geladen.")

        except Exception as e:
            print(f"Fout bij {tabelnaam}: {e}")

print("Full-Refresh voltooid.")
```
